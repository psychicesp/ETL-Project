```python
# Reference to CSV and reading CSV into Pandas DataFrame
csv_path = "../Full_Ramen.csv"
ramen_df = pd.read_csv(csv_path)
ramen_df.head(1)
```

```python
ramen_country_count_df = ramen_df["Country"].value_counts()
ramen_country_count_df
```

```python
#Country corrections
corrections = {
   'Country': {
      'USA': 'United States',
      'UK': 'United Kingdom',
      'Souh Korea': 'South Korea',
      'Dubai':'United Arab Emirates',
      'Sarawak':'Malaysia'}
}

ramen_df.replace(corrections, regex=True, inplace=True)

print(ramen_df)
```

```python
ramen_country_count_df = ramen_df["Country"].value_counts()
ramen_df.head()
```

```python
#Set Continent List Manually
North_America = ['United States','Canada','Mexico']
South_America = ['Brazil','Peru','Colombia']
Australia = ['Australia']
Oceania = ['Fiji','New Zealand']
Africa = ['Ghana','Nigeria']
Europe = ['Finland','France','Germany','Holland','Italy','Poland','Spain','Sweden','United Kingdom','Portugal','Hungary','Netherlands','Estonia','Ukraine']
Asia = ['Bangladesh','Cambodia','China','India','Indonesia','Israel','Japan','Malaysia','Myanmar','Nepal','Pakistan','Philippines','Phlippines','Singapore','South Korea','Taiwan','Thailand','Hong Kong','Russia','Vietnam']


# Add Continent column
ramen_df = ramen_df.copy()
ramen_df['Continent'] = ''

for i in ramen_df.index:
    
    if ramen_df['Country'][i] in North_America:
        ramen_df.loc[i,'Continent'] = 'North America'
        
    if ramen_df['Country'][i] in South_America:
        ramen_df.loc[i,'Continent'] = 'South America'
    
    if ramen_df['Country'][i] in Australia:
        ramen_df.loc[i,'Continent'] = 'Australia'
        
    if ramen_df['Country'][i] in Oceania:
        ramen_df.loc[i,'Continent'] = 'Oceania'
        
    if ramen_df['Country'][i] in Africa:
        ramen_df.loc[i,'Continent'] = 'Africa'
        
    if ramen_df['Country'][i] in Europe:
        ramen_df.loc[i,'Continent'] = 'Europe'
    
    if ramen_df['Country'][i] in Asia:
        ramen_df.loc[i,'Continent'] = 'Asia'
        
ramen_df
```
